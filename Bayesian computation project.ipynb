```python
def prior_normal(x, mean_beta, sigma_beta):
    
    n = len(x)
    denominator = pow(2*np.pi,n/2)*pow(np.linalg.det(sigma_beta),0.5)
    
    in_exp = -0.5*(x-mean_beta).transpose()@np.linalg.inv(sigma_beta)@(x-mean_beta)
    distribution = (1/denominator)*np.exp(in_exp)
    
    return distribution
```

```python
def likelihood_normal(y,data, beta, sigma):
    
    mean = data@beta
    n = len(mean)
    
    denominator = pow(2*np.pi,n/2)*pow(np.linalg.det(sigma),0.5)
    
    in_exp = -0.5*(y-mean).transpose()@np.linalg.inv(sigma)@(y-mean)
    distribution = (1/denominator)*np.exp(in_exp)
    
    return distribution
    
```

```python
def posterior_unormalized(x, y, data, beta, sigma, mean_beta, sigma_beta):
    
    return prior_normal(x, mean_beta, sigma_beta)*likelihood_normal(y,data, beta, sigma)
```

```python
def compute_gradient(x, y, data, beta, sigma, mean_beta, sigma_beta):
    
    gradient = - posterior_unormalized(x, y, data, beta, sigma, mean_beta, sigma_beta)*np.linalg.inv(sigma)@(y-data@beta)
    
    return gradient
```

To compute $\tilde{f}(F_1,...,F_d,Z_1,...,Z_d|data)$ we must compute :

- the prior $f(F_1,...,F_d,Z_1,...,Z_d)$
- the likelihood $f(data|F_1,...,F_d,Z_1,...,Z_d)$
